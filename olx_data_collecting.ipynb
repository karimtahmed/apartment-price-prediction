```python
import re
from collections import deque
import joblib
from math import ceil
from time import sleep
from itertools import product
from random import sample
from bs4 import BeautifulSoup
from selenium.common.exceptions import TimeoutException
from selenium.webdriver import Chrome, ChromeOptions
from selenium.webdriver.chrome.service import Service
from webdriver_manager.chrome import ChromeDriverManager

chrome_options = ChromeOptions()
prefs = {'profile.managed_default_content_settings.images': 2}
chrome_options.add_experimental_option('prefs', prefs)
chrome_installer = Service(ChromeDriverManager().install())
chrome = Chrome(service = chrome_installer, chrome_options = chrome_options)

max_sample_size = 64

try:
    
    ads, url_stack = joblib.load('state.pkl')

except:
    
    chrome.get('https://www.olx.com.eg/en/properties/apartments-duplex-for-sale/cairo/')
    sleep(2.0)
    soup = BeautifulSoup(chrome.page_source, 'html.parser')

    tags = soup.find_all('div', class_ = '_28a1a731')
    temps = [y.text for y in tags[1].find_all('a')]
    n_ads = [int(x) for x in re.findall('\d+', ', '.join(temps))]

    temps = [[y['href'] for y in x.find_all('a')] for x in tags if x.find_all('a')]
    temps = temps[1:]
    region_urls = ['https://www.olx.com.eg' + x for x in temps[0][2:]]
    temps = temps[1:]
    temps = [[y.split('=')[-1] for y in x] for x in temps]
    temps[0] = [temps[0][0]]
    del temps[1]

    temps[1] = temps[1][:2]
    temps[2] = temps[2][:3]
    temps[4] = temps[4][:1] + ['']
    temps[6] = temps[6][:2] + ['']
    temps[7] = temps[7][:1] + ['']
    temps[8] = temps[8][:2] + ['']
    del temps[3]
    del temps[4]

    search_configs = list(product(*temps[1:]))
    regions = zip(region_urls, n_ads)
    regions = sorted(regions, key = lambda x: x[1], reverse = True)
    max_n_ads = regions[0][1]
    regions = {x[0] : ceil(x[1] / max_n_ads * max_sample_size) if x[1] > 1960 else 1 for x in regions}
    url_stack = deque([f'{x}?sorting=asc-price&filter={temps[0][0]}' + ('%2C' + '%2C'.join(y) if regions[x] > 1 else '') for x in regions for y in sample(search_configs, regions[x])])

    ads = []

while url_stack:
    
    current_url = url_stack.popleft()
    
    scraped_urls = {x['url'] for x in ads}
    if current_url not in scraped_urls:
        
        print(len(ads), current_url)
    
        while True:
            try:
                chrome.get(current_url)
                #sleep(2.0)
                break
            except TimeoutException:
                print('timeout detected... cooling off...')
                sleep(2.0)
                
        soup = BeautifulSoup(chrome.page_source, 'html.parser')
        is_ad = current_url.split('/')[-2] == 'ad'
        if is_ad:

            ad = {'url': current_url}

            #continue scraping

            ad['title'] = soup.find('div', attrs = {'aria-label': 'Overview'}).h1.text.strip()

            ad['date'] = soup.find_all('span', attrs = {'aria-label': 'Creation date'})[-1].text.strip()
            ad['Location'] = soup.find_all('span', attrs = {'aria-label': 'Location'})[-1].text.strip()
            tag = soup.find('div', attrs = {'aria-label' : 'Details and description'})

            temp_tag = tag.div
            temp_dict = dict({tuple([y.text.strip() for y in x.find_all('span')]) for x in temp_tag.div.find_all('div')})

            tags = tag.find_all('div', class_ = temp_tag['class'])

            if tags:
                tags = tags[1:]

            if tags:
                key = tags[0].span.text.strip()
                value = tags[0].div.text.strip()
                temp_dict[key] = value
                tags = tags[1:]

            if tags:
                key = tags[0].span.text.strip()
                tags = tags[0].div.find_all('span')
                value = '; '.join([x.text.strip() for x in tags])
                temp_dict[key] = value

            ad.update(temp_dict)

            ##################

            ads.append(ad)
            state = ads, url_stack
            joblib.dump(state, 'state.pkl')

        else:

            tag = soup.find('div', role = 'navigation')
            try:
                if tag:
                    tags = tag.find_all('li')
                    tag = tags[-1].a
                    if tag:
                        nextpage_url = 'https://www.olx.com.eg' + tag['href']
                        url_stack.appendleft(nextpage_url)
            except:
                print("no pages")
                    
            tags = soup.find_all('li', attrs = {'aria-label': 'Listing'})
            ad_urls = {'https://www.olx.com.eg' + x.a['href'] for x in tags}
            url_stack.extendleft(ad_urls)
```
